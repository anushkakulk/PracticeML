```python
import numpy as np
import matplotlib.pyplot as plt
X  = np.array([0, 0, 1, 1, 2, 2, 3, 3])
y = np.array([0, 1, 1, 2, 2, 3, 3, 4])

X = np.vstack((X, y)).T

C = np.eye(8) - (1/8) * np.ones((8, 8)) # centering matrix

Q = np.dot(np.dot(X.T, C), X) # find the covariance matrix
[D, V] = np.linalg.eigh(Q)
print ("eigval, eigvector: ", D[0], V[0])
print ("eigval, eigvector: ", D[1], V[1]) 
# pick the max eigval, its associate eigvector is the answer! 

plt.plot(X[:, 0],  X[:, 0] * V[1][0] + V[1][1], label='Linear Model', color='blue')  # Plotting the projected vector
plt.scatter(X[:, 0], X[:, 1], label='X')  # Scatter plot of the original data points
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.show()
```
