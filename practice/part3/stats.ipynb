```python
import numpy as np
from scipy.stats import norm 
import matplotlib.pyplot as plt

## SAT 
X = np.genfromtxt('SAT.csv', delimiter='\n')
n, bins, patches = plt.hist(X, 20, facecolor ='blue', alpha = .7)
plt.title("SAT Scores")
plt.show()

print("average SAT score value: ", np.round(np.mean(X)))
mu, std = norm.fit(X)
probability_higher_than_1200 = 1 - norm.cdf(1200, loc=mu, scale=std)
print("probability of scoring higher than 1200:", 
       np.round(100 * probability_higher_than_1200, 2), "%")

## Lunch wait times
X = np.genfromtxt('lunch_wait_time.csv', delimiter='\n')
n, bins, patches = plt.hist(X, 20, facecolor ='blue', alpha = .7)
plt.title("Lunch Wait Times")
plt.show()
mu, std = norm.fit(X)
wait_longer_than_5 = 1 - norm.cdf(5, loc=mu, scale=std)
print("average lunch wait time value: ", np.mean(X))
print("probability of waiting longer than 5 mins:", 
       np.round(100  * wait_longer_than_5, 2), "%")


## Student age 
X = np.genfromtxt('student_age.csv', delimiter='\n')
n, bins, patches = plt.hist(X, 20, facecolor ='blue', alpha = .7)
plt.title("Student Ages")
plt.show()
mu, std = norm.fit(X)
younger_than_me = norm.cdf(19.58, loc=mu, scale=std)
print("average age value: ", np.mean(X))
print("probability of being younger than me (i'm 19.5):", 
       np.round(100  * younger_than_me, 2), "%")
```
